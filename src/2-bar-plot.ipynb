### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.


import matplotlib.pyplot as plt  

# Data  
categories = ['A', 'B', 'C', 'D']  
values = [5, 7, 3, 9]  
colors = ['#FF5733', '#33FF57', '#3357FF', '#F3FF33']  # Different colors for each bar  

# Create the bar plot  
plt.figure(figsize=(8, 5))  
plt.bar(categories, values, color=colors)  

# Customize the plot  
plt.title('Bar Plot of Categories vs Values', fontsize=16)  
plt.xlabel('Categories', fontsize=14)  
plt.ylabel('Values', fontsize=14)  

# Show the plot  
plt.tight_layout()  
plt.show()  
```
