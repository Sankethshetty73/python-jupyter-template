### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import numpy as np  
import matplotlib.pyplot as plt  

# Data  
x = np.linspace(0, 10, 100)  # For the line chart and scatter plot  
y_line = np.sin(x)            # Data for the line chart  
y_bar = [5, 7, 3, 9]          # Data for the bar chart  
y_scatter = np.random.rand(10) # Random data for the scatter plot  
labels = ['A', 'B', 'C', 'D']   # Labels for the bar chart  
sizes = [40, 25, 20, 15]        # Sizes for the pie chart  
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']  # Optional colors for pie chart  

# Create a 2x2 subplot layout  
fig, axs = plt.subplots(2, 2, figsize=(10, 10))  
fig.suptitle('Different Plots in Subplots', fontsize=16)  

# 1. Line Chart  
axs[0, 0].plot(x, y_line, label='y = sin(x)', color='blue')  
axs[0, 0].set_title('Line Chart')  
axs[0, 0].set_xlabel('X values')  
axs[0, 0].set_ylabel('Y values')  
axs[0, 0].legend()  
axs[0, 0].grid()  

# 2. Bar Chart  
axs[0, 1].bar(labels, y_bar, color='orange')  
axs[0, 1].set_title('Bar Chart')  
axs[0, 1].set_xlabel('Categories')  
axs[0, 1].set_ylabel('Values')  

# 3. Scatter Plot  
axs[1, 0].scatter(x[:10], y_scatter, color='green', s=100)  # Use only first 10 x values for scatter  
axs[1, 0].set_title('Scatter Plot')  
axs[1, 0].set_xlabel('X values')  
axs[1, 0].set_ylabel('Random Y values')  
axs[1, 0].grid()  

# 4. Pie Chart  
axs[1, 1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)  
axs[1, 1].set_title('Pie Chart')  

# Adjust layout  
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust for the main title  
plt.show()  
```
